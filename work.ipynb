```python
data = pd.read_csv('Electric_Vehicle_Population_Data.csv', sep=',', header=0)
data.head()
data.isnull().sum()
data = data.dropna()
data.isnull().sum()
```

```python
make_counts = data['Make'].value_counts()
print(make_counts)
```

```python
make_counts.plot(kind='bar')
plt.xlabel('Make')
plt.ylabel('Count')
plt.title('Number of Electric Vehicles by Make')
plt.tight_layout()
plt.show()
```

```python
utility_counts = data['Electric Utility'].value_counts()
print(utility_counts)
utility_counts.plot(kind='bar')
plt.xlabel('Electric Utility')
plt.ylabel('Count')
plt.title('Number of Electric Vehicles by Make')
plt.tight_layout()
plt.show()
```
